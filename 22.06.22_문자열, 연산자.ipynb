### 자주 사용하는 문자열 함수

```python
# count 함수 : 문자열에 포함된 문자 개수 세기
s = "오늘 점심 메뉴는 탕수육입니다. 저는 탕수육을 매우 좋아해서 탕수육만 먹을겁니다."
s.count('탕수육')
```

```python
# strip 함수 : 양쪽 공백 지우기
s = "  오늘 점심 메뉴는 탕수육입니다.  "
s
```

```python
s.strip()
```

```python
# replace 함수 : 특정문자를 원하는 문자로 변경하기
s = "오늘 점심 메뉴는 탕수육입니다. 저는 탕수육을 매우 좋아해서 탕수육만 먹을겁니다"
s.replace('탕수육', '마라탕')
```

```python
# split 함수 : 문자열을 특정 문자 기준으로 나눠주기
s.split(".")
```
