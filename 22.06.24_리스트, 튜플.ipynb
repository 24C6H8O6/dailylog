```python
"파이썬" not in str1
```

```python
list1 = [77,38,10]
33 in list1
```

```python
33 not in list1
```

```python
s = "Hi, My name is SeongWoo"
a = input("검색할 문자를 입력하세요 : ")
b = s.count(a)

if a in s :
    print(f"{a}는 {b}번 들어있네유!")
else :
    print(f"{a}는 들어있지 않네유...")
```
