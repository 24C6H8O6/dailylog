# KNN

- KNN : K-Nearest Neighbors(최근접이웃) => "거리가 가까운 것으로 예측"
    - 새로운 데이터 포인트와 가장 가까운 훈련 데이터셋의 데이터 포인트를 찾아 예측
    - 새로운 데이터로부터 거리를 구함

- KNeighborsClassifier(n_neighbors=이웃 수)
- 결론적으로 앙상블을 쓰게 됨(앙상블을 이해하기 위해 KNN,DecisionTree 배워야 함)
- 훈련 데이터 세트가 크면 (특성, 샘플의 수) 예측 느려짐
- 거리를 측정하기 때문에 단위를 맞춰줘야 하는 정규화 필요


# Decision Tree

- 지니 불순도 : 질문을 바꿔가면서 수치 파악
- 타깃 값이 한 개인 리프 노드를 순수 노드라고 한다.
- 모든 노드가 순수 노드가 될 때까지 학습하면 복잡해지고 과대적합이 된다.
- 새로운 데이터 포인트가 들어오면 해당하는 노드를 찾아 분류하면 더 많은 클래스를 선택하고 회귀라면 평균을 구한다.
