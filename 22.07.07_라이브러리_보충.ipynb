### 데이터 필터링(불리언 색인)
- 조건식이 들어감

```python
df['아파트'].unique()
```

```python
df[df['아파트']=='우미린에듀시티']
```

```python
# 아파트 중 호반현대 골라서 5층이상인 곳
df[(df['아파트']=='호반현대') & (df['층']>5)]     
```

```python
# 아파트 면적이 130이상인 것만 필터링 
df[df['면적'] >= 130]
```

```python
# 아파트 면적이 130이상인 것만 필터링
# 그중에 아파트 이름과 가격만 보고싶다.
df[df['면적']>=130][['아파트', '가격']]
```

```python
# 아파트 가격이 1.5억 미만인 아파트이름과 가격을 필터링
df[df['가격'] < 15000][['아파트','가격']]
```

### 데이터 가공

```python
# 면적과 가격을 이용해서 단가 컬럼을 생성하고 추가해보자.
df['단가'] = df['가격']/df['면적']
df.head()
```

```python
# 지역별 아파트 평균가격을 알아보자.
df['지역'].unique()
```

```python
def split_str(row):
    return row.split(" ")[0]
```

```python
df['가격'].mean()
```

```python
# apply(사용자 지정 함수)
df['지역2'] = df['지역'].apply(split_str)
```

```python
df.head()
```

```python
# 지역2를 이용해서 광주광역시의 평균 아파트 가격을 알아보자.
df[df['지역2']=="광주광역시"]["가격"].mean()
```

```python
df[df['지역2']=="서울특별시"]["가격"].mean()
```
