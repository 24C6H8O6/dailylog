```python
import numpy as np
```

### 사용할 데이터셋 생성

#### 문제 생성(공부한 시간)

```python
# 0.0~1.0 사이의 균등분포를 가진 100개 데이터가 랜덤하게 추출
x = np.random.uniform(low=0.0, high=1.0, size=100)
print(x.shape)
print(x)
```

#### 정답 생성(성적)

```python
y = 10 * x + 5
print(y.shape)
print(y)
```

### 경사하강법을 이용한 학습

```python
num_epoch = 500   # 경사하강법을 적용할 최대 반복 횟수
learning_rate = 1.0  # 학습률

# 초기 가중치와 절편을 셋팅
w = np.random.uniform(low=0.0, high=1.0)
b = np.random.uniform(low=0.0, high=1.0)

print("초기 가중치 : ", w)
print("초기 절편 : ", b)
```

```python
for epoch in range(num_epoch):
    # MSE 구하기
    pre_y = x * w + b
    mse = ((pre_y - y)**2).mean()
    print("MSE : ",mse)
    
    # MSE가 일정 수준이하로 떨어지면 반복을 멈춤
    if mse < 0.001 :
        break
        
    # 경사하강법을 이용해 w,b 업데이트
    w = w - learning_rate * ((pre_y-y)*x).mean()
    b = b - learning_rate * ((pre_y-y)*1).mean()
    
    # 중간결과 출력
    print('epoch :',(epoch+1),'w :',w,'b :',b)
```

### 입력특성이 2개인 경우 구현

#### 문제 생성(x1 : 공부한시간, x2 : 잠자는 시간)

```python
# 0.0~1.0 사이의 균등분포를 가진 100개 데이터가 랜덤하게 추출
x1 = np.random.uniform(low=0.0, high=1.0, size=100)
x2 = np.random.uniform(low=0.0, high=1.0, size=100)
```

#### 정답 생성(성적)

```python
y = 13 * x1 + 5 * x2 - 2
```

#### 학습

```python
num_epoch = 500   # 경사하강법을 적용할 최대 반복 횟수
learning_rate = 1.0  # 학습률

# 초기 가중치와 절편을 셋팅
w1 = np.random.uniform(low=0.0, high=1.0)
w2 = np.random.uniform(low=0.0, high=1.0)
b = np.random.uniform(low=0.0, high=1.0)

print("초기 가중치 : ", w1, w2)
print("초기 절편 : ", b)
```

```python
for epoch in range(num_epoch):
    # MSE 구하기
    pre_y = w1*x1 + w2*x2 + b
    mse = ((pre_y - y)**2).mean()
    print("MSE : ",mse)
    
    # MSE가 일정 수준이하로 떨어지면 반복을 멈춤
    if mse < 0.001 :
        break
        
    # 경사하강법을 이용해 w,b 업데이트
    w1 = w1 - learning_rate * ((pre_y-y)*x1).mean()
    w2 = w2 - learning_rate * ((pre_y-y)*x2).mean()
    b = b - learning_rate * ((pre_y-y)*1).mean()
    
    # 중간결과 출력
    print('epoch :',(epoch+1),'w1 :',w1,'w2 :',w2,'b :',b)
```
