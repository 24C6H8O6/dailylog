#### numpy 불리언 인덱싱
- 특정 조건에 맞으면 true, 아니면 False 값으로 배열을 만들어 조건에 맞는 값에 접근하는 방법

```python
# 50~99까지의 랜덤한 8개 값을 가지는 1차원 배열 생성
arr20 = np.array(np.random.randint(50,100, size=8))
arr20
```

```python
# 짝수인 값들을 True로 출력시켜보세요~!
arr20 % 2 == 0
```

```python
# True, False 값이 아닌 실제값을 보고 싶을 경우
# 배열명[배열 내에서 찾고자 하는 조건]
arr20[arr20 % 2 == 0]
```

### 예제) BMI지수 및 과체중 BMI값 확인

#### 1. 데이터 로드

```python
# delimiter(구분자) : 파일에서 데이터를 구분해주는 문자 지정
data = np.loadtxt('data/height_weight.txt', delimiter=',')
data

# 키는 cm, 몸무게는 kg
```

#### 2. 데이터의 속성확인
- 배열 형태
- 배열 차원수
- 배열 전체요소 개수

```python
print("배열 형태 : ", data.shape)
print("배열 차원수 : ", data.ndim)
print("배열 전체 요소 개수 : ", data.size)
```

#### 데이터에서 키와 몸무게 데이터들을 각각 변수에 담기

```python
# 키
height = data[0]
# 몸무게
weight = data[1]
```

```python
height
```

```python
weight
```

```python
bmi = weight / ((height/100)**2)
bmi
```

#### 4. BMI지수가 23이상 25이하인 사람은 몇 명일까요?
- 불리언 인덱싱에서는 and→&, or→|, not→~를 쓴다.

```python
# BMI가 23이상 25이하인 값 먼저 출력
# 출력된 값의 개수 구하는 명령 사용
bmi[(bmi >= 23) & (bmi <= 25)].size     #  len(bmi[(bmi >= 23) & (bmi <= 25)])과 동일함
```

```python
# 23이상이거나 19이하
bmi[(bmi >= 23) | (bmi <= 19)].size
```

#### 4. 1번 사용자가 매긴 평점의 평균 구하기
- 사용자 번호의 유일한 값(유니크 값) 확인
- 사용자 번호의 유니크 값 개수 확인
- 1번 사용자의 데이터에 접근(불리언 인덱싱)
- 1번 사용자가 매긴 평점들의 평균 구하기

```python
# 평점을 매긴 사용자가 몇 명인지 알아보자!
# unique :  유일한 값을 구해주는 명령
user_id = np.unique(movie[:, 0])
user_id
```

```python
# 사용자는 총 6040명
user_id.size
```

```python
# 사용자의 법호가 1번인 데이터의 경우
a= movie[movie[:, 0] == 1]
a
```

```python
np.mean(a[:, 2])
```

```python
a[:, 0].size
```

```python
# 1번 사용자가 매긴 평점들
array = a[:,2]
array
```

```python
np.mean(array)
```

- 1번 사용자는 전체 영화의 평점 평균보다 후하게 평점을 줬구나~!

#### 5. 각 사용자가 매긴 평점들의 평균 구하기!

```python
movie
```

```python
movie[movie[:,0]==1]
```

```python
movie[movie[:,0]==1][:,2]
```

```python
movie[movie[:,0]==1][:,2].mean()
```

```python
user_mean_list = []

for i in user_id:        # range(1,6041,1)
    # 1번~6040번 사용자가 매긴 평점들의 평균을 각각 구해주는 코드
    user_mean = movie[movie[:,0]==i][:,2].mean()
    # 빈 리스트에 사용자 번호와 그 사람이 매긴 평점의 평균을 리스트로 저장
    user_mean_list.append([i,user_mean])

user_mean_list
```

#### 6. 위에서 구한 각 사용자별 평점 평균이 4점 이상인 사용자만 구해보세요~!
- 사용자 번호와 명수를 구해봅시다!

```python
user_mean_arr = np.array(user_mean_list, dtype='int64')
user_mean_arr
```

```python
user_mean_arr[user_mean_arr[:, 1] >= 4][:, 0].size
```

- 사용자가 매긴 평점의 평균이 4점 이상인 사람들은 총 1544명
