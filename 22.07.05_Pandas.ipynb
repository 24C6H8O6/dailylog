### Pandas 란?
데이터 처리와 분석에 최적화된 라이브러리
- 행과 열(테이블 형태)로 이루어진 데이터 객체를 만들어 다룰 수 있음
- 빅데이터를 처리하고 분석하는데 편리한 함수들을 제공

### Pandas 구조
- Series : 1차원 데이터들의 집합(index + value)
- DataFrame : 2차원의 행과 열로 이루어진 표 형태의 데이터 집합(1차원의 Series가 모여서 DF를 이룸)

### Pandas 실습 예제!!!
2015~2017년 광주광역시 범죄현황 데이터를 이용해 전년 대비 지역별 범죄 증감율을 구해보자!
- (증감률 공식 = (금년-작년)/작년*100)

#### 데이터 로드
- 인덱스 컬럼은 '관서명'으로 설정할 것
- 인코딩은 'euc-kr'로 설정할 것

```python
# 내 방법
# 1. 파일 불러와서 다른 DF와 다른 값인 광주지방경찰청제거
# 2. 각 DF의 '구분' 열 제거
# 3. 각 DF의 '총합' 열 추가
# 4. 각 DF의 총계부분 행만 구별
# 5. 증감율 계산
# 6. concat() 으로 행 합치기
# 7. columns 의 제목 정하기
```

```python
df_2015 = pd.read_csv('data/2015.csv', encoding='euc-kr', index_col='관서명')
df_2016 = pd.read_csv('data/2016.csv', encoding='euc-kr', index_col='관서명')
df_2017 = pd.read_csv('data/2017.csv', encoding='euc-kr', index_col='관서명')
```

```python
df_2017.drop("광주지방경찰청", axis=0, inplace=True)
df_2017

del df_2015['구분']
df_2015['2015총계'] = df_2015.sum(axis=1)
df_2015_total = df_2015[::6]['2015총계']
del df_2016['구분']
df_2016['2016총계'] = df_2016.sum(axis=1)
df_2016_total = df_2016[::6]['2016총계']
del df_2017['구분']
df_2017['2017총계'] = df_2017.sum(axis=1)
df_2017_total = df_2017[::6]['2017총계']
df_15_16 = (df_2016_total - df_2015_total)/df_2015_total*100
df_16_17 = (df_2017_total - df_2016_total)/df_2016_total*100
```

```python
total = pd.concat([df_2015_total, df_15_16, df_2016_total, df_16_17, df_2017_total], axis=1)
```

```python
total.columns = ['2015총계','2015~16증감율','2016총계','2016~17증감율','2017총계']
total
```

```python
# 쌤 방법
# 1. 파일을 불러온 상태에서 파일의 형태(shape) 일일이 확인
# 2. 나머지 2개에는 없는 광주지방경찰청 제거
# 3. 각 DF에 sum값 더하기
# 4. 불리언 인덱싱을 활용하여 '구분' 컬럼의 '발생건수'의 '총합'에 종합하기
# 5. 증감율 구하기
# 6. concat 사용하여 'columns'들 합치기
# 7. 칼럼 제목 정하기
# 8. 출력
```

```python
df_2015 = pd.read_csv('data/2015.csv', encoding='euc-kr', index_col='관서명')
df_2016 = pd.read_csv('data/2016.csv', encoding='euc-kr', index_col='관서명')
df_2017 = pd.read_csv('data/2017.csv', encoding='euc-kr', index_col='관서명')
```

```python
print(df_2015.shape)
print(df_2016.shape)
print(df_2017.shape)
```

```python
# 광주지방경찰청 제거
df_2017.drop("광주지방경찰청", axis=0, inplace=True)
df_2017
```

```python
df_2015["총합"] = df_2015.sum(axis=1)
df_2016["총합"] = df_2016.sum(axis=1)
df_2017["총합"] = df_2017.sum(axis=1)
```

```python
# 불리언 인덱싱을 활용하여 '구분' 컬럼의 '발생건수'의 '총합'에 종합해보자!
s15 = df_2015.loc[df_2015['구분']=='발생건수', '총합']
print(s15)
s16 = df_2016.loc[df_2016['구분']=='발생건수', '총합']
print(s16)
s17 = df_2017.loc[df_2017['구분']=='발생건수', '총합']
print(s17)
```

```python
# 증감율 구하기!
# 증감율 공식 : (금년-작년)/작년*100
```

```python
s1516 = (s16-s15)/s15*100
s1617 = (s17-s16)/s16*100
```

```python
s1516
```

```python
total = pd.concat([s15,s1516,s16,s1617,s17], axis=1)
```
