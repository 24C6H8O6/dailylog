```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# pyplot은 matplotlib을 간단하고 쉽게 표현해주는 모듈
# 복잡하고 다양하게 표현하는 pylab도 있으나 pyplot에 비해 많은 정보를 입력해야 함
```

### plot차트 : 선 그래프

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y)   # 하나의 라인
plt.show()       # 그려진 라인들을 하나의 표로 표기
```

- Line style

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.')
plt.show()
```

- Line width

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2)
plt.show()
```

- Marker

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2, marker='o')
plt.show()
```

- Line color

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x, y, ls='-.', lw=2, marker='o', c='c')
plt.show()
```

#### 차트 옵션

- 그림 범위 지정

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(-2,10)   # x축의 범위
plt.ylim(-3,12)   # y축의 범위

plt.show()
```

- 틱(tick) 설정
  - 차트나 그래프에서 축상의 위치 표시 지점을 틱(tick)이라고 함
  - 틱에 쓰여준 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
  - 틱의 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동으로 설정하고 싶다면 xticks, yticks 명령을 사용!

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.show()
```

- 그리드(grid)
  - 그림의 틱 위치를 잘 보여주기 위해 차트 내부에 격자 선을 그려주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

#### 사고유형이 차대차 사건 중에 죽거나 다친 사람이 많은 발생지시도를 알아보고 시각화해보자!

- 중요한 키워드들이 들어있는 위치 확인
  - 사고유형_대분류 에서 '차대차'인 행만 출력
  - '차대차'인 행 중에서 '발생지시도'와 '사상자수'컬럼 확인

```python
data['사고유형_대분류'].value_counts()
```

```python
ac = data[data['사고유형_대분류']=='차대차']
ac
```

```python
ac[['발생지시도', '사상자수']]
```

```python
# 사고유형이 차대차인 사건의 총 개수
len(ac)
```

```python
# 발생한 1642개의 사건 중에서 시도 각각의 개수를 확인
ac['발생지시도'].value_counts()
```

#### 각 발생지시도별로 사상자수의 합계 구하기

```python
ac2 = ac[['발생지시도', '사상자수']].groupby('발생지시도').sum()
ac2
```

```python
ac2.index
```

```python
len(ac2.index)
```

### 예제) 차vs차 교통사고 사상자 수 시각화하기

- x, y축 값 설정

```python
# x축에 들어갈 bar의 개수는 17개
x = np.arange(len(ac2.index))

# y축은 위에서 구한 발생지시도별 사상자수 데이터를 내림차순 정렬로 넣어줌
y = ac2['사상자수'].sort_values(ascending=False)
```

```python
x
```

```python
y.index
```

- 그래프 그리기

```python
# 차트 크기 변경(가로, 세로)
plt.figure(figsize=(13,7))

plt.bar(x, y)

plt.xticks(x, y.index)
plt.xlabel("지역")
plt.ylabel("사상자 수")
plt.title("차vs차 교통사고의 사상자 수")

plt.show()
```
